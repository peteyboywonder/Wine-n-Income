```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the datasets
wine_df = pd.read_csv('wine.csv')
adult_df = pd.read_csv('adult.csv')

# Identify categorical columns
wine_cat_columns = wine_df.select_dtypes(include='object').columns
adult_cat_columns = adult_df.select_dtypes(include='object').columns

# Delete duplicate rows
wine_df = wine_df.drop_duplicates()
adult_df = adult_df.drop_duplicates()

# Identify and address missing values
wine_df = wine_df.fillna(method='ffill')  # You can use different techniques like mean, median or mode
adult_df = adult_df.fillna(method='ffill')

# Identify and correct inconsistencies in data for categorical values
for cat_col in wine_cat_columns:
    wine_df[cat_col] = wine_df[cat_col].str.lower().str.strip().replace('cats', 'cat')

for cat_col in adult_cat_columns:
    adult_df[cat_col] = adult_df[cat_col].str.lower().str.strip().replace('cats', 'cat')

# Produce univariate visuals for the target and all features
for column in wine_df.columns:
    sns.histplot(data=wine_df, x=column)
    plt.show()

for column in adult_df.columns:
    sns.histplot(data=adult_df, x=column)
    plt.show()

# Identify outliers (using IQR method, you can use Z-score or any other method as well)
def identify_outliers_iqr(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    return (df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))

wine_outliers = identify_outliers_iqr(wine_df)
adult_outliers = identify_outliers_iqr(adult_df)

# Print outliers (you may decide to remove or modify them)
print("Wine dataset outliers:\n", wine_outliers)
print("Adult dataset outliers:\n", adult_outliers)

# Save the cleaned data
wine_df.to_csv('cleaned_wine.csv', index=False)
adult_df.to_csv('cleaned_adult.csv', index=False)
```
